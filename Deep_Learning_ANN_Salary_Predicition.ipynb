```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('/content/drive/MyDrive/data sets 2/salary.csv')
```

```python
data
```

```python
data.isna().sum()
```

```python
data=data.drop('marital-status', axis=1)
data=data.drop('relationship', axis=1)
data=data.drop('race', axis=1)
data=data.drop('sex', axis=1)
data=data.drop('education-num', axis=1)
```

```python
data
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
data['workclass']=le.fit_transform(data['workclass'])
data['education']=le.fit_transform(data['education'])
data['occupation']=le.fit_transform(data['occupation'])
data['native-country']=le.fit_transform(data['native-country'])
```

```python
data
```

```python
x=data.drop('salary', axis=1)
y=data.salary
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.3, random_state=0)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python

# Creating a pipeline
model=Sequential()
```

```python
# 1st hidden layer with input layer
model.add(Dense(units=145,activation='relu',input_dim=13))
```

```python
# 2nd hidden layer
model.add(Dense(units=120,activation='relu',))
```

```python
# 3rd hidden layer
model.add(Dense(units=70,activation='relu',))
```

```python
# output layer
model.add(Dense(units=1,activation='sigmoid'))
```

```python
model.summary()
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model_his=model.fit(x_train,y_train,validation_split=0.20, batch_size=55,epochs=25,verbose=1)
```
